```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
URL="https://ticker.finology.in/"
HEADERS=({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36','Accept-Language':'en-US, en;q=0.5'})
```

```python
webpage=requests.get(URL,headers=HEADERS)
```

```python
a=BeautifulSoup(webpage.content,"html.parser")
print(a)
```

```python
q=a.find("h1")
print(q.text)
```

```python
w=a.find("a",attrs={'class':'btn btn-link'})
print(w)
```

```python
c=a.find_all("a",attrs={'class':'nav-link'})
print(c)
for i in c:
  print(i.text)
```

```python
soup=BeautifulSoup(webpage.text,"lxml")
table=soup.find("div",class_="col-12 col-md-6")
print(table)
```

```python
headers=table.find_all("th")
print(headers)

titles=[]
for i in headers:
  x=i.text
  titles.append(x)
print(titles)
```

```python
rows=table.find_all("a")
rowss=[]
for i in rows:
  y=i.text
  rowss.append(y)
print(rowss)
c=rowss[2::]
print(c)
print(len(c))
```

```python
price=table.find_all("td")
prices=[]
for i in price:
  z=i.text
  prices.append(z)
print(prices)
```

```python
price=table.find_all("td",attrs={"class":'Number'})
prices=[]
for i in price:
  z=i.text
  prices.append(z)
print(prices)
print(len(prices))
```

```python
m=[]
n=[]
for i in range(0,len(prices)):
  if(i%2==0):
    m.append(prices[i])
  else:
    n.append(prices[i])
print(m)
print(len(m))
print(n)
print(len(n))
```

```python
m=table.find_all("tr")
for i in m[1:]:
  print(i.text)
```

```python
headers = [cell.get_text(strip=True) for cell in m[0].find_all(["th", "td"])]

# Extract data from subsequent rows
data = [
    [cell.get_text(strip=True) for cell in row.find_all(["td", "th"])]
    for row in m[1:]
]

# Create a DataFrame
df = pd.DataFrame(data, columns=headers)

# Save the DataFrame to a CSV file
df.to_csv("output.csv", index=False)
```
